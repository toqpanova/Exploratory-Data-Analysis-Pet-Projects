18. Создайте функцию, где смотря столбец 'Electric Range', необходимо заполнить новый столбец 'Electric Range Mapped' по условиям, если 'Electric Range' меньше 100 то 'Electric Range Mapped' равно 'Low', от 100 до 250 'Medium', выше 250 'High'

```python
def Electric_Range_Map(value):
  if value < 100:
    return 'Low'
  elif 100 <= value <= 250:
    return 'Medium'
  else:
    return 'High'

df['Electric Range Mapped'] = df['Electric Range'].map(Electric_Range_Map)
```

```python
df[['Electric Range','Electric Range Mapped']]
```

```python
df[df['Electric Range Mapped'] == 'High']
```
