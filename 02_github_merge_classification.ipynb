–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'Python' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Full stack'.

–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'PHP' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Full stack'.

–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'Java' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Backend'.

–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'Go' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Backend'.

–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'HTML' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Frontend'.

–ï—Å–ª–∏ 'language' —Å–æ–¥–µ—Ä–∂–∏—Ç 'CSS' , —Ç–æ 'web development' –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å–ª–æ–≤–æ 'Frontend'.

–ê –≤–æ –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞—è—Ö –ø—É—Å—Ç—ã–µ —è—á–µ–π–∫–∏

```python
def full_column(value):
  if value ==  'JavaScript':
    return 'Frontend'
  elif value ==  'Python':
    return 'Full stack'
  elif value ==  'PHP':
    return 'Full stack'
  elif value ==  'Java':
    return 'Backend'
  elif value ==  'Go':
    return 'Backend'
  elif value ==  'HTML':
    return 'Frontend'
  elif value ==  'CSS':
    return 'Frontend'
  else:
    return 'NaN'

sorted_df['web development'] = sorted_df['language'].map(full_column)
```

–ï—Å–ª–∏ –∫—Ç–æ-—Ç–æ –Ω–∏–∂–µ –∏–¥–µ–∞–ª—å–Ω–æ–≥–æ –≤–µ—Å–∞ –¥–ª—è —Å–≤–æ–µ–≥–æ —Ä–æ—Å—Ç–∞, –æ–Ω ¬´Thin¬ª(—Ç–æ –µ—Å—Ç—å –•—É–¥–æ–π). –ï—Å–ª–∏ –∫—Ç–æ-—Ç–æ –≤—ã—à–µ –∏–¥–µ–∞–ª—å–Ω–æ–≥–æ –≤–µ—Å–∞ –¥–ª—è —Å–≤–æ–µ–≥–æ —Ä–æ—Å—Ç–∞, –æ–Ω ¬´Obese¬ª (—Ç–æ –µ—Å—Ç—å –û–∂–∏—Ä–µ–≤—à–∏–π). –ï—Å–ª–∏ –∫—Ç–æ-—Ç–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –∏–¥–µ–∞–ª—å–Ω–æ–≥–æ –≤–µ—Å–∞ –¥–ª—è —Å–≤–æ–µ–≥–æ —Ä–æ—Å—Ç–∞, –æ–Ω  ¬´Fit¬ª (—Ç–æ –µ—Å—Ç—å –ü–æ–¥—Ç—è–Ω—É—Ç—ã–π).

–°–º–æ—Ç—Ä—è –Ω–∞ —ç—Ç–∏ –ø—Ä–∞–≤–∏–ª–∞ –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø–æ–ª–Ω–∏—Ç –ø–æ —ç—Ç–∏–º –∫—Ä–∏—Ç–µ—Ä–∏—è–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü "Body Type" –≤ –Ω–∏–∂–µ—É–∫–∞–∑–∞–Ω–Ω–æ–º –¥–∞—Ç–∞ —Ñ—Ä–µ–π–º–µ

```python
import pandas as pd
df = pd.DataFrame(
         {
             'Name': ['Edward', 'Natalie', 'Chris M', 'Priyatham'],
             'Sex' : ['M', 'F', 'M', 'M'],
             'Age': [45, 35, 29, 26],
             'weight(kgs)': [68.4, 58.2, 64.3, 53.1],
             'height(cms)' : [178, 160, 173, 168]
         }
     )
```

```python
male_fitness = {
     #height  : (weight_lower_cap, weight_upper_cap)
     178      : (     67.5       ,      83         ),
     173      : (     63         ,      70.6       ),
     168      : (     58         ,      70.7       )
 }
female_fitness = {
     #height  : (weight_lower_cap, weight_upper_cap)
     160      : (     47.2       ,      57.6       )
 }
```

```python
from re import X
def body_type(row):
    if row['Sex'] == 'M':
        x = male_fitness.get(row['height(cms)'], (None, None))
    elif row['Sex'] == 'F':
        x = female_fitness.get(row['height(cms)'], (None, None))
    else:
        x = (None, None)

    if x != (None, None):
        min_weight, max_weight = x
        if row['weight(kgs)'] < min_weight:
            return 'Thin'
        elif row['weight(kgs)'] > max_weight:
            return 'Obese'
        else:
            return 'Fit'
    else:
        return 'Unknown'


df['Body Type'] = df.apply(body_type, axis=1)

df

```
