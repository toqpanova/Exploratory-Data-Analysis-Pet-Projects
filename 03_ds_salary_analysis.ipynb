13. Наидите среднюю зарплату ('salary_in_usd') по должностям ('job_title')

```python
avg_salary_by_job = data.groupby('job_title')['salary_in_usd'].mean()
print(avg_salary_by_job)
```

14. Постройте график найденные коэффициенты по средним зарплатам ('salary_in_usd') по должностям ('job_title')

```python
avg_salary_by_job = avg_salary_by_job.sort_values(by='salary_in_usd', ascending=False)
plt.figure(figsize=(10, 8))
plt.barh(avg_salary_by_job['job_title'], avg_salary_by_job['salary_in_usd'])
plt.title('Средняя зарплата по должностям')
plt.xlabel('Средняя зарплата (USD)')
plt.ylabel('Должности')
plt.gca().invert_yaxis()

plt.show()
```

15. Необходимо найти те специальности ('job_title') которые имели самый выскокий рост по медианной зарплате ('salary_in_usd') в процентном соотношение к пред. году по медианной зарплате

```python
median_salary = data.groupby(['job_title', 'work_year'])['salary_in_usd'].median().reset_index()

median_salary['salary_pct_change'] = (
    median_salary.groupby('job_title')['salary_in_usd']
    .pct_change() * 100
)
median_salary['salary_pct_change'] = median_salary['salary_pct_change'].round(6)

sorted_salary_change = median_salary.sort_values(
    by='salary_pct_change', ascending=False
).dropna()

growth = sorted_salary_change[['work_year', 'job_title', 'salary_in_usd', 'salary_pct_change']]
growth
```
